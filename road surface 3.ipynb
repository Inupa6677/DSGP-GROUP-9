```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('C:/Users/Vihanga Ranaweera/Documents/GitHub/DSGP-GROUP-9/Birmingham_road_surface_dataset.csv')
df
```

```python
new_df = df.drop(columns=['Year'])
new_df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
new_df['Zone'] = label_encoder.fit_transform(df['Zone'])
new_df['Day of the Week'] = label_encoder.fit_transform(df['Day of the Week'])
new_df['Time Period'] = label_encoder.fit_transform(df['Time Period'])
new_df['Road Surface'] = label_encoder.fit_transform(df['Road Surface'])
new_df
```

```python
new_df = new_df.dropna()
new_df
```

```python
# Iterate over unique road surface conditions
road_surface_conditions = new_df['Road Surface'].unique()

for condition in road_surface_conditions:
    # Filter data for the current road surface condition
    condition_data = new_df[new_df['Road Surface'] == condition]
    
    train_x, test_x, train_y, test_y = train_test_split( condition_data[['Day of the Week', 'Time Period']], condition_data['Acc Count'], test_size=0.2, random_state=42 )

    model = RandomForestClassifier()
    model.fit(train_x, train_y)

# def road_surface_accuracy(day_of_week, time_period, model):
#     input_data = [[day_of_week, time_period]]
#     prediction = model.predict(input_data)
#     return prediction[0]

    predictions = model.predict(test_x)
    
#     user_input_day_of_week = 0
#     user_input_time_period = 8 

#     final_road_surface_accuracy = road_surface_accuracy(user_input_day_of_week, user_input_time_period, model)

    accuracy = accuracy_score(test_y, predictions)
    print(f'Accuracy for Road Surface {condition}: {accuracy}')
```
