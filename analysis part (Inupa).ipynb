```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_to_analyze = pd.read_csv('C:/Users/MSI KATANA/Desktop/DSGP Git Repo/DSGP-GROUP-9/edit1Newdataset.csv')
df_to_analyze
```

```python
df_to_analyze.info() 
```

```python
# Checking  missing values in the DataFrame
missing_values_in_data_frame = df_to_analyze.isnull()

# number of missing values in each column
missing_values_count_in_data_frame = df_to_analyze.isnull().sum()

# print the result
print("Missing Values in Each Column:")
print(missing_values_count_in_data_frame)
```

```python
# drop the rows with any missing values
df_to_analyze = df_to_analyze.dropna()
df_to_analyze
```

```python
# remove the duplicate values in rows 
df_to_analyze = df_to_analyze.drop_duplicates()
df_to_analyze
```

```python
# Checking  missing values in the DataFrame
missing_values_in_data_frame = df_to_analyze.isnull()

# number of missing values in each column
missing_values_count_in_data_frame = df_to_analyze.isnull().sum()

# print the result
print("Missing Values in Each Column:")
print(missing_values_count_in_data_frame)
```

```python
# check for unique values for each column
unique_values_per_column = df_to_analyze.apply(lambda x: x.unique())

print(unique_values_per_column)
```

```python
# accidents in different weather condition
weather_counts = df_to_analyze['Weather_Conditions'].value_counts()
ax = weather_counts.plot(kind='barh', color='skyblue')

plt.title('Accident Distribution by Weather Conditions')
plt.xlabel('Accident Count')
plt.ylabel('Weather Conditions')
plt.grid(axis='x')

# to get the count value on the top
for i, v in enumerate(weather_counts):
    ax.text(v + 3, i, str(v), color='black', va='center')  

plt.tight_layout()

plt.show()
```

```python
# Box plots for the distribution of accident counts in (urban vs. rural) areas
plt.figure(figsize=(8, 6))
sns.boxplot(x='Urban / Rural', y='Number_of_Vehicles', data=df_to_analyze, palette='pastel')
plt.title('Accident Counts in Urban vs. Rural Areas')
plt.xlabel('Urban / Rural')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Frequency of accidents for different road types
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Type', data=df_to_analyze, palette='Set2')
plt.title('Accident Frequency by Road Type')
plt.xlabel('Road Type')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()

# Frequency of accidents for different road surface conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Surface', data=df_to_analyze, palette='pastel')
plt.title('Accident Frequency by Road Surface Condition')
plt.xlabel('Road Surface Condition')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()

# Frequency of accidents for different speed limits
plt.figure(figsize=(10, 6))
sns.countplot(x='Speed_limit', data=df_to_analyze, palette='viridis')
plt.title('Accident Frequency by Speed Limit')
plt.xlabel('Speed Limit (mph)')
plt.ylabel('Accident Count')
plt.show()

# Relationship between the number of vehicles involved and accident severity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number_of_Vehicles', y='Accident_Severity', data=df_to_analyze, hue='Accident_Severity', palette='Dark2')
plt.title('Relationship between Number of Vehicles and Accident Severity')
plt.xlabel('Number of Vehicles Involved')
plt.ylabel('Accident Severity')
plt.legend(title='Accident Severity')
plt.show()
```

```python
# Compare accident counts for different light conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Light_Conditions', data=df_to_analyze, palette='twilight_shifted')
plt.title('Accident Counts by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of accident severity under various lighting conditions
light_severity_counts = pd.crosstab(df_to_analyze['Light_Conditions'], df_to_analyze['Accident_Severity'])
plt.figure(figsize=(12, 6))
light_severity_counts.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Distribution of Accident Severity by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Accident Count')
plt.legend(title='Accident Severity')
plt.xticks(rotation=45)
plt.show()
```
