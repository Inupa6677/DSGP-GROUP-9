```python
import pandas as pd
```

```python
# Load the first dataset
df1 = pd.read_csv('UK_Accident.csv')  

# Load the second dataset
df2 = pd.read_csv('newDataset.csv') 
print(df1)
print(df2) 
```

```python
# Select two attributes from each dataset
new_df1 = df1[["Accident_Index", "Location_Easting_OSGR", "Location_Northing_OSGR", "Longitude", "Latitude"]]
new_df2 = df2[["Local_Authority_(District)", "Weather_Conditions", "Number_of_Vehicles", "Junction_Detail"]]

# Combine the selected attributes into a new dataset
new_dataset = pd.concat([new_df1, new_df2], axis=1)

# Display the new dataset
print(new_dataset)
```

```python
print(new_dataset.head())        # Print the first few rows
print(new_dataset.info())        # Overview of dataset
print(new_dataset.describe())    # Statistical summary for numerical columns
```

```python
# To remove rows with missing values
new_dataset.dropna(inplace=True)

print(new_dataset)
```

```python
# To remove duplicate rows
new_dataset.drop_duplicates(inplace=True)
print(new_dataset)
```

```python
# Optionally, reset the index after dropping rows
new_dataset.reset_index(drop=True, inplace=True)
print(new_dataset)
```

```python
# Save the cleaned dataset to a new CSV file, if needed
new_dataset.to_csv('cleaned_dataset.csv', index=False)
```
