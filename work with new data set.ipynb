```python
# first i loaded the dataset
df1 = pd.read_csv('C:/Users/MSI KATANA/Desktop/DSGP Git Repo/DSGP-GROUP-9/edit1Newdataset.csv')
df1
```

```python
# main selected fields
selected_columns = ['Accident_Index', 'Date', 'Day_of_Week', 'Time', 'Local_Authority_(District)'] 
df_selected = df1[selected_columns]
df_selected
```

```python
# Checking  missing values in the DataFrame
missing_values_in_data_frame = df_selected.isnull()

# number of missing values in each column
missing_values_count_in_data_frame = df_selected.isnull().sum()

# print the result
print("Missing Values in Each Column:")
print(missing_values_count_in_data_frame)
```

```python
# drop the rows with any missing values
df_selected = df_selected.dropna()
df_selected
```

```python
# remove the duplicate values in rows 
df_selected = df_selected.drop_duplicates()
df_selected
```

```python
# Checking  missing values in the DataFrame
missing_values_in_data_frame = df_selected.isnull()

# number of missing values in each column
missing_values_count_in_data_frame = df_selected.isnull().sum()

# print the result
print("Missing Values in Each Column:")
print(missing_values_count_in_data_frame)
```

```python
# confirmd no missing values
```

```python
# Assuming your original DataFrame is named 'df_selected'
# Specify the correct date format using the 'format' parameter and 'dayfirst=True'
df_selected['Year'] = pd.to_datetime(df_selected['Date'], format='%m/%d/%Y', dayfirst=True).dt.year

# Map day of the week numbers to day names
day_mapping = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}
df_selected['Day_of_the_Week'] = pd.to_datetime(df_selected['Date'], format='%m/%d/%Y', dayfirst=True).dt.dayofweek.map(day_mapping)

# Create a new column for time periods (1 to 24)
df_selected['Time_Period'] = pd.to_datetime(df_selected['Time']).dt.hour + 1

# Create a new DataFrame with the desired format
result_df = df_selected.groupby(['Local_Authority_(District)', 'Year', 'Day_of_the_Week', 'Time_Period'])['Accident_Index'].count().reset_index()
result_df.rename(columns={'Accident_Index': 'Acc. Count'}, inplace=True)

# Print the first few rows of the result DataFrame
print(result_df.head(1000))
```

```python
# Check the current data type of 'Date' column
print(df['Date'].dtype)

# If it's not datetime, convert it
df['Date'] = pd.to_datetime(df['Date'])

# Check the data type again
print(df['Date'].dtype)

# Reset the index before creating the time series plot
df_reset_index = df.groupby('Date').size().reset_index(name='Accident Count')

# Create a time series plot
plt.figure(figsize=(14, 6))
sns.lineplot(x='Date', y='Accident Count', data=df_reset_index, color='black')

plt.title('Accidents Over Time')
plt.xlabel('Date')
plt.ylabel('Accident Count')
plt.show()
```

```python
# Convert 'Time_Period' to string type
result_df['Time_Period'] = result_df['Time_Period'].astype(str)

# Extract the numerical part of 'Time_Period'
result_df['Time_Period_Num'] = result_df['Time_Period'].str.extract('(\d+)').astype(int)

# Sort DataFrame by 'Time_Period_Num' for proper line plotting
result_df_sorted = result_df.sort_values(by='Time_Period_Num')

# Create a line plot for aggregated 'Acc. Count' over different time periods
plt.figure(figsize=(14, 6))
sns.lineplot(x='Time_Period', y='Acc. Count', data=result_df_sorted, marker='o')

plt.title('Accident Count Over Different Time Periods')
plt.xlabel('Time Period')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_columns = result_df.select_dtypes(include=['float64', 'int64']).columns

# Compute the correlation matrix
correlation_matrix = result_df[numerical_columns].corr()

# Create a heatmap for visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
from scipy import stats  

# Box plot to visualize outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x='Acc. Count', data=result_df)
plt.title('Box Plot of Acc. Count')
plt.show()

# Calculate z-scores for 'Acc. Count'
z_scores = stats.zscore(result_df['Acc. Count'])

# Define a threshold for z-scores, e.g., 3
threshold = 3

# Identify and remove outliers
result_df_no_outliers = result_df[(z_scores < threshold) & (z_scores > -threshold)]

# Display the DataFrame without outliers
print(result_df_no_outliers.head())
```
